```python
# 导入训练数据集
titanic = pd.read_csv('data/train.csv')
# 导入测试数据集
test_data = pd.read_csv('data/test.csv')
```

```python
# 数据清洗与预处理
# 填充训练集缺失值
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
titanic['Fare'] = titanic['Fare'].fillna(titanic['Fare'].median())
titanic['Embarked'] = titanic['Embarked'].fillna(titanic['Embarked'].mode()[0])
titanic.head()
# 填充测试集缺失值
test_data['Age'] = test_data['Age'].fillna(titanic['Age'].median())
test_data['Fare'] = test_data['Fare'].fillna(titanic['Fare'].median())
test_data['Embarked'] = test_data['Embarked'].fillna(test_data['Embarked'].mode()[0])

# 性别编码
titanic['Sex'] = titanic['Sex'].map({'male':0, 'female':1})
test_data['Sex'] = test_data['Sex'].map({'male':0, 'female':1})

# 登船港口编码
titanic['Embarked'] = titanic['Embarked'].map({'S':0, 'C':1, 'Q':2})
test_data['Embarked'] = test_data['Embarked'].map({'S':0, 'C':1, 'Q':2})
```

```python
# 数据清洗与预处理

# 设置 PassengerId 为索引
titanic.set_index('PassengerId', inplace=True)

# 删除缺失值过多的 Cabin 数据
titanic.drop('Cabin', axis=1, inplace=True)

# 年龄缺失值处理（使用中位数填充空缺值）
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())

# 性别数值化
titanic['Sex'] = titanic['Sex'].map({'male': 0, 'female': 1})

# 登船港口缺失值填充 
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
# 登船港口数值化
titanic['Embarked'] = titanic['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# 票价缺失值填充
titanic['Fare'].fillna(titanic['Fare'].median(), inplace=True)

titanic.head()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold

# 属性间相关性
df = titanic[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']]
cor = df.corr()
print(cor)

# 热力图
sns.heatmap(cor, annot=True, cmap='coolwarm')
plt.show()

# 预测变量
predictors = ['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']

# 逻辑回归模型
alg = LogisticRegression(max_iter=200)
kf = KFold(n_splits=3, shuffle=True, random_state=1)

predictions = []

# KFold 交叉验证
for train_index, test_index in kf.split(titanic):
    train_predictors = titanic[predictors].iloc[train_index, :]
    train_target = titanic['Survived'].iloc[train_index]
    
    alg.fit(train_predictors, train_target)
    
    test_predictions = alg.predict(titanic[predictors].iloc[test_index, :])
    predictions.append(test_predictions)

# 合并预测结果
predictions = np.concatenate(predictions)
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib

# ---- 设置中文字体，解决绘图中文显示问题 ----
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False





# ---- 3️⃣ 特征与目标 ----
predictors = ['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
target = 'Survived'

# ---- 4️⃣ 训练集划分 ----
X_train, X_valid, y_train, y_valid = train_test_split(
    titanic[predictors], titanic[target], test_size=0.3, random_state=42)

# 构建决策树模型 (CART) ----
tree_model = DecisionTreeClassifier(
    criterion='entropy',  # 信息熵划分
    max_depth=4,
    random_state=42
)

# 训练模型
tree_model.fit(X_train, y_train)

# ---- 6️⃣ 验证模型 ----
y_pred = tree_model.predict(X_valid)
acc = accuracy_score(y_valid, y_pred)
print(f"验证集准确率: {acc:.4f}")

# ---- 7️⃣ 特征重要性 ----
importance = tree_model.feature_importances_
for f, imp in zip(predictors, importance):
    print(f"{f}: {imp:.4f}")

# 可视化特征重要性
sns.barplot(x=importance, y=predictors)
plt.title("特征重要性")
plt.show()

# ---- 8️⃣ 绘制决策树 ----
plt.figure(figsize=(18,8))
plot_tree(tree_model,
          feature_names=predictors,
          class_names=['Not Survived', 'Survived'],
          filled=True,
          rounded=True)
plt.title("Titanic CART决策树")
plt.show()

# 测试集预测 ----
X_test_extra = test_data[predictors]
test_predictions = tree_model.predict(X_test_extra)


```
